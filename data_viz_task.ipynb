# Data visualisation task 

Using data visualisaiton techniques to analyse the Cars93 dataset.

**Additional information**  
[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
print(df)
```

### 1. Boxplot

```python
# Boxplot for revs per mile
manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
selected_manufacturers = df[df['Manufacturer'].isin(manufacturers)]
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data= selected_manufacturers)
plt.title('Rev per Mile by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Revs per Mile')
plt.show() 
```

**As the box plot above suggests, Suzuki has the car with the highest revs per mile.**


### 2. Histogram

```python
# Histogram of MPG
# I have done two separate histograms, one for MPG City and one for MPG Highway
x = df['MPG.city']
y = df['MPG.highway']    
plt.hist(x, color='blue', label='MPG City')
plt.hist(y, color='green', label='MPG Highway')
plt.title('Fuel Efficiency: City VS Highway')
plt.legend()
# I have included the legend() function too display the labels for MPG City and MPG Highway
plt.show()
```

**According to the histograms, cars tend to run more miles per gallon on highways than in the city, so it is generally more fuel efficeint to drive on highways.**

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
sns.lineplot(x='Wheelbase', y='Turn.circle', data=df)
plt.show()
```

**As shown in the lineplot, there is a general positive correlation between turn cirvle and wheelbase, as the wheelbase gets larger, turn circle increases too, but it seems to plateu towards the end when wheelbase reaches around 116-117.**


### 4. Barplot

```python
# Barplot for mean horsepower
group_by_type = df.groupby(by=['Type'])
# Firstly, I am grouping the data by car type
car_data_avg = round(group_by_type.mean(numeric_only=True), 0) 
car_data_avg.sample(5)
# Here, I am calculating the mean value of all columns for each car type
selected_features = pd.DataFrame({'Car Type': car_data_avg.index,
                                  'Mean Horsepower': car_data_avg['Horsepower']})
#'Type' is the index of the selected_features dataframe
# which means it can be accessed through the '.index' function
selected_features.plot(kind='bar', ylabel='Average Horsepower')
plt.title('Mean Horsepower for Each Car Type')
```

**From the bar chart above, it seems that going from small to midsize to large, the mean horsepower does tend to increase. However, if Compact cars are smaller than Small cars and Vans are bigger than Large cars, then it means that large cars do not necessarily have more horesepower since Compact cars have a higher mean horsepower than Small cars and Vans have a lower average horsepower than Large cars. It is also worth noting that Sporty cars have a relatively high mean horsepower but their size is ambiguous. This can also help support the argument that larger cars do not necessarily have more horsepower.**
